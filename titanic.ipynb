##### Handling Missing Values (Data Cleaning)

```python
data.isna().sum()
```

```python
# data['Embarked'].mode()
data['Embarked'].fillna((data['Embarked'].mode()[0]), inplace=True)
```

```python
data.isna().sum()
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
data.isna().sum()
```

```python
data.drop('Cabin', axis=1, inplace=True)
data.isna().sum()
```

```python
data.duplicated().sum() 
```
